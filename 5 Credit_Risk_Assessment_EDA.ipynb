## Dealing with the columns having outliers

#### YEARS_EMPLOYED COLUMN

```python
df1.AGE.max()
```

>**As the maximum age in the dataset is 69.1 so the maximum years employed cannot be more than 55.1 (69.1-14)<br>
Because the minimum age for employment in India is 14 years**

```python
#dropping values where employment years is more than 55.1 
df1.YEARS_EMPLOYED=df1.YEARS_EMPLOYED.apply(lambda x:x if x<55.1 else np.nan)
```

```python
sns.boxplot(df1.YEARS_EMPLOYED);
```

#### AMT_INCOME_TOTAL COLUMN

```python
df1.AMT_INCOME_TOTAL.describe()
sns.boxplot(df1.AMT_INCOME_TOTAL);
```

- There is a outlier in AMT_INCOME_TOTAL and its value is extremely high, So we can remove this value for better analysis.
- We can remove the value above 99th percentile

```python
df1=df1[df1.AMT_INCOME_TOTAL<np.percentile(df1.AMT_INCOME_TOTAL,99)]
```

```python
sns.boxplot(df1.AMT_INCOME_TOTAL);
```

- After removing the value above 99th percentile it can be observed that AMT_INCOME is now fairly distributed.
